```python
import pandas
```

```python
#Leitura dos arquivos
df1 = pandas.read_excel("Aracaju.xlsx")
df2 = pandas.read_excel('Fortaleza.xlsx')
df3 = pandas.read_excel('Natal.xlsx')
df4 = pandas.read_excel('Recife.xlsx')
df5 = pandas.read_excel('Salvador.xlsx')
```

```python
df5.head()
```

```python
#Juntando todos os arquivos
df = pandas.concat([df1,df2,df3,df4,df5])
```

```python
df.head()
```

```python
df.tail()
```

```python
df.sample(5)
```

```python
df.dtypes
```

```python
df['LojaID'] = df['LojaID'].astype('object')
```

```python
df.dtypes
```

```python
df.head()
```

```python
# Consultando linhas com valores nulos
df.isnull().sum()
```

```python
# Substituindo os valores nulos pela média
df['Vendas'].fillna(df['Vendas'].mean(), inplace=True)
```

```python
df['Vendas'].mean()
```

```python
df.isnull().sum()
```

```python
df.sample(15)
```

```python
# Substituindo os valores por zero
df['Vendas'].fillna(0, inplace=True)
```

```python
# Apagando os valores nulos
df.dropna(inplace=True)
```

```python
# Apagando linhas com valores nulos com base apenas em 1 coluna
df.dropna(subset=['Vendas'], inplace=True)
```

```python
# Removendo linhas com valroes nulos em todas as colunas
df.dropna(how='all', inplace=True)
```

Criando colunas novas

```python
# Criando colunas novas
df['Receita'] = df['Vendas'].mul(df['Qtde'])
```

```python
df.head()
```

```python
# Criando uma coluna com base na Receita e Vendas
df['Receita/Vendas'] = df['Receita'] / df['Vendas']
```

```python
df.head()
```

```python
# Retornando maior receita
df['Receita'].max()
```

```python
# Retornando menor receita
df['Receita'].min()
```

```python
# nlargest Retornando coluna com maior número
df.nlargest(3, 'Receita')
```

```python
# nsmallest Retornando coluna com menor número
df.nsmallest(3, 'Receita')
```

```python
# Agrupando Uma coluna com outra
df.groupby('Cidade')['Receita'].sum()
```

```python
# Ordenando conjunto de dados do maior para o menos com base no head()
df.sort_values('Receita', ascending=False).head(10)
```

Trabalhando com Datas

```python
# Transformando uma coluna em um inteiro
df['Data'] = df['Data'].astype('int64')
```

```python
# Verificando o tipo de dado de cada coluna
df.dtypes
```

```python
df['Data'] = pandas.to_datetime(df['Data'])
```

```python
df.dtypes
```

```python
# Agrupamento por ano
df.groupby(df['Data'].dt.year)['Receita'].sum()
```

```python
# Criando uma nova coluna com o ano
df['Ano_Venda'] = df['Data'].dt.year
```

```python
df.sample(5)
```

```python
# Extraindo o mês e o dia da venda
df['mes_venda'], df['dia_venda'] = (df['Data'].dt.month, df['Data'].dt.day)
```

```python
df.sample(5)
```

```python
# Retornando a data mais antiga
df['Data'].min()
```

```python
# Calculando a diferença de dias
df['diferenca_de_dia'] = df['Data'] - df['Data'].min()
```

```python
df.sample(5)
```

```python
# Criando coluna de trimestre
df['trimestre_venda'] = df['Data'].dt.quarter
```

```python
df.sample(5)
```

```python
df.dtypes
```

```python
# Filtrando vendas
vendas_marco_19 = df.loc[(df['Data'].dt.year == 2019) & (df['Data'].dt.month == 3)]
```

```python
vendas_marco_19
```
